## 3.1 Homographies

교재 본문 정의 : A homography is a 2D projective transformation that maps points in one plane to another. In our case, the planes are images or planar surfaces in 3D.

즉, Homography는 3D를 평면으로 변환할때 (픽셀) 점의 변환 되는 것을 의미하고 아래의 수식에서 3x3의 행렬을 일반적으로 지칭합니다.

\begin{equation*}
\begin{bmatrix}
x' \\ y' \\ w'
\end{bmatrix}
=
\begin{bmatrix}
h_{1} & h_{2} & h_{3} \\ h_{4} & h_{5}& h_{6} \\ h_{7} & h_{8}& h_{9} 
\end{bmatrix}
\begin{bmatrix}
x \\ y \\ w
\end{bmatrix} 
\end{equation*}

\begin{equation*}
\mathbf{x}'=H\mathbf{x}
\end{equation*}

여기서 $x$, $y$는 이미지의 좌표를 의미하고, $w$는 이미지의 기울어진 정도를 의미합니다.
Homogeneous coordinates에서의 점 좌표는 다음과 같이 정의 합니다.
$\mathbf{x}=[x, y, w]=[\alpha x,\alpha y,\alpha w]=[x/w,y/w,1]$ 이로 인해 Homography $\mathbf{H}$는 각도(기울어진 정도)에 대해 독립적이게 됩니다.
그래서 앞의 $\mathbf{H}$가 쉬워지게 됩니다.

다음에 정의하는 함수는 $\mathbf{H}$를 간단하게 만들어 주는 역할을 합니다.

```python
def normalize(points):
    """Normalize a collection of points in
     homogeneous coordinates so that last row = 1."""
    for row in points:
        row /= points[-1]
    return points

def make_homog(points):
    """Convert a set of points (dim*n array) to
    homogeneous coordinates."""
    return vstack((points, ones((1, points.shape[1]))))
```

즉, 앞의 matrixs 수식을 다음과 같이 간단하게 정리할 수 있습니다.

\begin{equation}
\begin{bmatrix}
x' \\ y' \\ 1
\end{bmatrix}
=
\begin{bmatrix}
a_{1} & a_{2} & t_{x} \\ a_{3} & a_{4}& t_{y} \\ 0 & 0& 1 
\end{bmatrix}
\begin{bmatrix}
x \\ y \\ 1
\end{bmatrix} 
\end{equation}

이 수식에서 의미하는 $\mathbf{H}$는 좌표변환을 의미하므로 affine transformation이라고도 합니다. 또한 아래와 같이 표현할 수 있습니다.

\begin{equation}
\mathbf x'
=
\begin{bmatrix}
\mathbf A & \mathbf t \\ 0 & 1 
\end{bmatrix}
\mathbf x
\end{equation}

 affine transformation matrix $\mathbf H$는 에서 matrix $\mathbf A$는 회전행렬(an invertible matrix), vector $\mathbf t =[t_{x}, t_{y}]$는 평행이동(translation)을 나타냅니다.
 Affine transformation을 사용한 예로는 warping이 있습니다.

A similarity transformation

\begin{equation}
\begin{bmatrix}
x' \\ y' \\ 1
\end{bmatrix}
=
\begin{bmatrix}
s \cos(\theta) & -s \sin(\theta) & t_{x} \\ s \sin(\theta) & s \cos(\theta)& t_{y} \\ 0 & 0& 1 
\end{bmatrix}
\begin{bmatrix}
x \\ y \\ 1
\end{bmatrix} 
\end{equation}

\begin{equation}
\mathbf x'
=
\begin{bmatrix}
s\mathbf R & \mathbf t\\ 0 & 1 
\end{bmatrix}
\mathbf x 
\end{equation}

(물리학의) 입자 혹은 그래픽스를 공부하신적이 있다면, 위의 Matrixs가 친숙할 겁니다.
 
 $\mathbf R$은 $\theta$에 따른 회전을 의미하며, $\mathbf t$는 평행이동을 의미합니다. 거리를 의미하는 $s$가 1로 보존될 경우를 강체변환(a rigid transformation)이라 합니다.
 예로 영상정합(image registration)에서 비슷한 변환이 사용됩니다.

### The Direct Linear Transformation Algorithm

 Homographies는 직접적으로 두 이미지(혹은 두 평면)을 계산할 수 있습니다.
 좀 더 쉽게 설명하자면, 모든 투영변환이 갖는 8 degrees의 제약이 없어집니다.
  $\mathbf H$를 알기 위해서는 $\mathbf H$관계가 point간 관계이고, 이를 x와 y좌표를 가진 두 수식으로 표현됩니다. 두 수식의 미지수를 알기 위해서는 최소한 (실세계에서 동일한) 4개의 points를 알고 있어야 합니다(앞의 $\mathbf H$참조).
  
 The direct lenear transformation(DLT)는 $\mathbf H$를 계산하기 위한 알고리즘 입니다.
 
 (이 알고리즘은 교재와 함께 제공되는 모듈에 구현되어 있습니다)
 
 최소자승법(a least square solution)으로 $\mathbf H$의 singular value decomposition(SVD)를 찾습니다.

# 3.2 Warping Images

Warping은 요약하면, 이미지를 기하학적인 변화를 주는 것입니다.
An affine transformation matrix $\mathbf H$는 warping(affine warping)으로 불리고 3D 컴퓨터 그래픽스 분야에서 자주 사용됩니다.

구현하는 코드 형태는 다음과 같습니다.
transformed_im = ndimage.affine_transform(im,A,b,size)

```python
from matplotlib.pyplot import *
from PIL import Image
from scipy import ndimage
from numpy import *

im = array(Image.open('data/empire.jpg').convert('L'))
H = array([[1.4, 0.05, -100], [0.05,1.5,-100],[0,0,1]])

#affine_transform 호출
im2 = ndimage.affine_transform(im,H[:2,:2], (H[0,2],H[1,2]))

#결과출력
%matplotlib inline
figure()
gray()
imshow(im)
show()
imshow(im2)
show()
```

### Piecewise Affine Warping

제목대로 부분적으로 Affine Warping을 하는 내용을 다룹니다. 즉, Affine Warping을 작은 부위로 나누어서 한다는 의미입니다.

아래의 코드는 이를 보여주기 위한 예제입니다.

```python
import matplotlib.delaunay as md

x,y = array(random.standard_normal((2,100)))
centers,edges,tri,neighbors = md.delaunay(x,y)

figure()
for t in tri:
    t_ext = [t[0],t[1],t[2],t[0]]  # add first point to end
    plot(x[t_ext],y[t_ext],'r')

plot(x,y,'*')
axis('off')
show()
```

```python
from pylab import *
from numpy import *
from PIL import Image

# PCV는 폴더명임
from PCV.geometry import homography, warp

"""
This is the piecewise affine warp example from Section 3.2, Figure 3-5.
"""

# open image to warp
fromim = array(Image.open('data/sunset_tree.jpg')) 
x,y = meshgrid(range(5),range(6))

x = (fromim.shape[1]/4) * x.flatten()
y = (fromim.shape[0]/5) * y.flatten()

# triangulate
tri = warp.triangulate_points(x,y)

# open image and destination points
im = array(Image.open('data/turningtorso1.jpg'))
tp = loadtxt('data/turningtorso1_points.txt', 'int') # destination points

# convert points to hom. coordinates (make sure they are of type int)
fp = array( vstack((y,x,ones((1,len(x))))), 'int')
tp = array( vstack((tp[:,1],tp[:,0],ones((1,len(tp))))), 'int')

# warp triangles
im = warp.pw_affine(fromim,im,fp,tp,tri)

# plot
figure()
imshow(im) 
warp.plot_mesh(tp[1],tp[0],tri) 
axis('off')
show()
```

## 3.3 Creating Panoramas

### RANSAC

"RANdom SAmple Consensus"의 약자가 "RANSAC"입니다.

이름대로 샘플링을 하는 방법으로 랜덤으로 하여 오차가 작은 것을 채택하는 방식입니다.

 RANSAC알고리즘에 대해 추가 참조(http://en.wikipedia.org/wiki/RANSAC)

### Robust Homography Estimation

실제 파노라마 이미지를 만들때 많이 사용되는 방법입니다.

 SIFT와 같이 특징점(feature point)에 RANSAC을 결합한 방법을 통해 $\mathbf H$를 얻는 방법입니다. 이렇게 얻은 $\mathbf H$를 이용해 warping을 하여 파노라마를 만들면, 단순 정합보다 자연스러운 이미지를 얻을 수 있습니다.
 
 (현재 예제파일에 '.shif'파일이 없어 시현을 할 수 없는 상태입니다. sift를 정확하게 정합하기 위해서는 많은 시간이 소요되기 때문에 여기서는 구현 코드만 적어 놓겠습니다.)

```python
from pylab import *
from numpy import *
from PIL import Image

# If you have PCV installed, these imports should work
from PCV.geometry import homography, warp
from PCV.localdescriptors import sift

"""
This is the panorama example from section 3.3.
"""

# set paths to data folder
featname = ['data/Univ'+str(i+1)+'.sift' for i in range(5)] 
imname = ['data/Univ'+str(i+1)+'.jpg' for i in range(5)]

# extract features and match
l = {}
d = {}
for i in range(5): 
    sift.process_image(imname[i],featname[i])
    l[i],d[i] = sift.read_features_from_file(featname[i])

matches = {}
for i in range(4):
    matches[i] = sift.match(d[i+1],d[i])

# visualize the matches (Figure 3-11 in the book)
for i in range(4):
    im1 = array(Image.open(imname[i]))
    im2 = array(Image.open(imname[i+1]))
    figure()
    sift.plot_matches(im2,im1,l[i+1],l[i],matches[i],show_below=True)


# function to convert the matches to hom. points
def convert_points(j):
    ndx = matches[j].nonzero()[0]
    fp = homography.make_homog(l[j+1][ndx,:2].T) 
    ndx2 = [int(matches[j][i]) for i in ndx]
    tp = homography.make_homog(l[j][ndx2,:2].T) 
    
    # switch x and y - TODO this should move elsewhere
    fp = vstack([fp[1],fp[0],fp[2]])
    tp = vstack([tp[1],tp[0],tp[2]])
    return fp,tp


# estimate the homographies
model = homography.RansacModel() 

fp,tp = convert_points(1)
H_12 = homography.H_from_ransac(fp,tp,model)[0] #im 1 to 2 

fp,tp = convert_points(0)
H_01 = homography.H_from_ransac(fp,tp,model)[0] #im 0 to 1 

tp,fp = convert_points(2) #NB: reverse order
H_32 = homography.H_from_ransac(fp,tp,model)[0] #im 3 to 2 

tp,fp = convert_points(3) #NB: reverse order
H_43 = homography.H_from_ransac(fp,tp,model)[0] #im 4 to 3    


# warp the images
delta = 2000 # for padding and translation

im1 = array(Image.open(imname[1]), "uint8")
im2 = array(Image.open(imname[2]), "uint8")
im_12 = warp.panorama(H_12,im1,im2,delta,delta)

im1 = array(Image.open(imname[0]), "f")
im_02 = warp.panorama(dot(H_12,H_01),im1,im_12,delta,delta)

im1 = array(Image.open(imname[3]), "f")
im_32 = warp.panorama(H_32,im1,im_02,delta,delta)

im1 = array(Image.open(imname[4]), "f")
im_42 = warp.panorama(dot(H_32,H_43),im1,im_32,delta,2*delta)


figure()
imshow(array(im_42, "uint8"))
axis('off')
show()
```

### Stitching the Image Together

 앞의 파노라마는 회전된 영상을 정합하여 기하학 적으로 왜곡된 부분이 있었습니다. 하지만, Stitching은 기하학적인 왜곡까지 고려하여 여러이미지를 정합하는 방법입니다.
 
 파노라마와 비교를 하자면, 파노라마는 한쪽 방향성(보통 가로방향)이 있는 반면, Stitching은 방향성 없이 정합하는 것으로 구분할 수 도 있습니다. MS에서는 인터넷에 있는 건축물 사진들을 통해 Stitching 한 Image를 토대로 3D모델링을 자동으로 해주는 프로그램을 시현하기 했었습니다.
